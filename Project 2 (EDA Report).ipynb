## Import Libraries

```python
# Importing required libraries

import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

## Load Dataset

```python
# Loading the cleaned dataset

df = pd.read_excel("Cleaned_E-Commerce_Project_2_Dataset.xlsx")

# Display first 5 rows
df.head()
```

## Dataset Overview

```python
# Shape of dataset

print("Rows and Columns:", df.shape)

# Column names
print("\nColumns in Dataset:\n")
print(df.columns)
```

## Data Information

```python
# Dataset information

df.info()
```

## Checking Missing Values

```python
# Missing values check

df.isnull().sum()
```

## Statistical Summary

```python
# Statistical summary of numerical columns

df.describe()
```

## Mean, Median and Count

```python
# Mean values

print("Mean Values:\n")
print(df.mean(numeric_only=True))

# Median values
print("\nMedian Values:\n")
print(df.median(numeric_only=True))

# Count values
print("\nCount Values:\n")
print(df.count())
```

## Checking Unique Values

```python
# Unique values in categorical columns

categorical_columns = df.select_dtypes(include='object').columns

for col in categorical_columns:
    print(f"\nUnique values in {col}:")
    print(df[col].unique())
```

## Total Revenue

```python
# Total Revenue

total_revenue = df["TotalPrice"].sum()

print("Total Revenue:", total_revenue)
```

## Average Order Value

```python
# Average Order Value

average_order_value = df["TotalPrice"].mean()

print("Average Order Value:", average_order_value)
```

## Most Sold Products

```python
# Top selling products

top_products = df["Product"].value_counts()

print(top_products)
```

## Visualization of Top Products

```python
# Top products bar chart

plt.figure(figsize=(10,6))

sns.countplot(data=df, y="Product", order=df["Product"].value_counts().index)

plt.title("Top Selling Products")
plt.xlabel("Count")
plt.ylabel("Products")

plt.show()
```

## Payment Method Analysis

```python
# Payment methods

payment_method = df["PaymentMethod"].value_counts()

print(payment_method)
```

## Payment Method Visualization

```python
# Payment method visualization

plt.figure(figsize=(7,7))

df["PaymentMethod"].value_counts().plot.pie(autopct='%1.1f%%')

plt.title("Payment Method Distribution")

plt.ylabel("")

plt.show()
```

## Order Status Analysis

```python
# Order status count

order_status = df["OrderStatus"].value_counts()

print(order_status)
```

## Order Status Visualization

```python
# Order status chart

plt.figure(figsize=(8,5))

sns.countplot(data=df, x="OrderStatus")

plt.title("Order Status Analysis")

plt.show()
```

## Referral Source Analysis

```python
# Referral source analysis

referral_source = df["ReferralSource"].value_counts()

print(referral_source)
```

## Referral Source Visualization

```python
# Referral source visualization

plt.figure(figsize=(8,5))

sns.countplot(data=df, x="ReferralSource")

plt.title("Referral Source Distribution")

plt.xticks(rotation=45)

plt.show()
```

## Distribution of Total Price

```python
# Distribution plot

plt.figure(figsize=(10,5))

sns.histplot(df["TotalPrice"], bins=30, kde=True)

plt.title("Distribution of Total Price")

plt.show()
```

## Distribution of Quantity

```python
# Quantity distribution

plt.figure(figsize=(10,5))

sns.histplot(df["Quantity"], bins=20, kde=True)

plt.title("Distribution of Quantity")

plt.show()
```

## Outlier Detection (Boxplot)

```python
# Boxplot for TotalPrice

plt.figure(figsize=(10,5))

sns.boxplot(x=df["TotalPrice"])

plt.title("Outlier Detection - TotalPrice")

plt.show()
```

## Correlation Analysis

```python
# Correlation matrix

correlation = df.corr(numeric_only=True)

print(correlation)
```

## Correlation Heatmap

```python
# Heatmap visualization

plt.figure(figsize=(8,6))

sns.heatmap(correlation, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

## Monthly Sales Trend

```python
# Convert Date column to datetime

df["Date"] = pd.to_datetime(df["Date"])

# Extract month

df["Month"] = df["Date"].dt.month

# Monthly sales

monthly_sales = df.groupby("Month")["TotalPrice"].sum()

print(monthly_sales)
```

## Monthly Sales Visualization

```python
# Monthly sales trend chart

plt.figure(figsize=(10,5))

monthly_sales.plot(marker='o')

plt.title("Monthly Sales Trend")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.grid()

plt.show()
```
